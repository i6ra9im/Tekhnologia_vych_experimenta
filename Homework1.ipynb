```python
x = np.linspace(0, 100, 50)                 
 y = (1/(1+np.sin(x)+4*np.cos(x))) 
 plt.style.use('seaborn')                        
fig, ax = plt.subplots()  
ax.plot(x, y, 'r', label="y(x)")  
ax.legend()     
plt.show()
```

```python
a=min(y)
b=max(y)
c=(a+b)/2
print('min=', a)
print('max=', b)
print('average=', c)
```

```python
def diff_arr(x, y):
    len_arr = len(y)
    if len(x) != len_arr:
        return None
    array_diff = np.zeros(len_arr)
    for i in range(len_arr-1):
        array_diff[i] = (y[i+1] - y[i])/(x[i+1]-x[i])
    array_diff[len_arr-1] = array_diff[len_arr - 2]
    return array_diff
    
Dif = diff_arr(x,y)

plt.style.use('seaborn')    
fig, ax = plt.subplots()
ax.plot(x, Dif, linewidth=3)
plt.show() 
```
